## Cleaning the column 'value'

```python
df['Value']=df['Value'].str.replace('�','')
```

```python
def cleanFunc(x):
    if x[-1] == 'M':
        x = x[:-1]+'*1000000'
        return eval(x)
    elif x[-1] == 'K':
        x = x[:-1]+'*1000'
        return eval(x)
    
```

```python
df['Value'] = df['Value'].apply(cleanFunc)
```

## Cleaning column 'Wage'

```python
df['Wage']=df['Wage'].str.replace('�','')
```

## Cleaning column 'Release Clause'

```python
def replace_Joined(x):
    if int(x[-2:])<=22:
        return '20' + x[-2:]
    else:
        return '19' + x[-2:]
```
